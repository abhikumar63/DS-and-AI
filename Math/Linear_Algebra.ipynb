# **Linear Algebra**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define 3D Vectors: [Action, Comedy, Romance] (Scale 0-10)
matrix = np.array([10, 2, 1])     # High Action
john_wick = np.array([9, 1, 0])   # Very High Action
the_holiday = np.array([1, 5, 10]) # High Romance/Comedy

print(f"Matrix Vector: {matrix}")
print(f"John Wick Vector: {john_wick}")
print(f"The Holiday Vector: {the_holiday}")
```

## Vector Addition and Scalar Multiplication

```python
# 1. Scalar Multiplication: Scaling the "intensity" of The Matrix
scaled_matrix = 2 * matrix

# 2. Vector Addition: Creating a "Double Feature" vector
double_feature = matrix + the_holiday

# Visualization (2D Projection for clarity)
plt.figure(figsize=(8, 6))
origin = np.array([[0, 0, 0], [0, 0, 0]]) # origin point

# We plot only the first two dimensions [Action, Comedy] for visualization
plt.quiver([0, 0, 0], [0, 0, 0],
           [matrix[0], the_holiday[0], double_feature[0]],
           [matrix[1], the_holiday[1], double_feature[1]],
           color=['r', 'b', 'g'], scale=25, label=['Matrix', 'The Holiday', 'Combined'])

plt.title("Vector Addition: The 'Parallelogram' Result")
plt.xlabel("Action Score")
plt.ylabel("Comedy Score")
plt.legend()
plt.grid()
plt.show()
```

## The Dot Product (Similarity)

```python
# The Dot Product calculation
dot_similar = np.dot(matrix, john_wick)
dot_different = np.dot(matrix, the_holiday)

print(f"Dot Product (Matrix & John Wick): {dot_similar}")
print(f"Dot Product (Matrix & The Holiday): {dot_different}")

# Conclusion: Higher value = Higher Similarity in features.
```

## Norms (Magnitude)

```python
# Calculate L2 Norm (Euclidean Distance from origin)
norm_matrix = np.linalg.norm(matrix)
norm_holiday = np.linalg.norm(the_holiday)

# Unit Vector: Direction without magnitude
unit_matrix = matrix / norm_matrix

print(f"Magnitude of Matrix profile: {norm_matrix:.2f}")
print(f"Unit Vector (Direction Only): {unit_matrix}")
```

## Cosine Similarity

```python
def cosine_similarity(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

sim_1 = cosine_similarity(matrix, john_wick)
sim_2 = cosine_similarity(matrix, the_holiday)

print(f"Cosine Similarity (Matrix vs John Wick): {sim_1:.4f}") # Near 1.0 (Identical direction)
print(f"Cosine Similarity (Matrix vs The Holiday): {sim_2:.4f}") # Near 0.0 (Orthogonal/Different)
```
