## Defining the "Transformation" Function

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_transformation(matrix, title="Linear Transformation"):
    # Create a grid of points (a square)
    x, y = np.meshgrid(np.linspace(0, 1, 10), np.linspace(0, 1, 10))
    points = np.vstack([x.flatten(), y.flatten()])

    # Apply transformation: T = A * P
    transformed_points = matrix @ points

    # Plotting
    plt.figure(figsize=(6, 6))
    plt.scatter(points[0], points[1], color='gray', alpha=0.3, label='Original')
    plt.scatter(transformed_points[0], transformed_points[1], color='blue', label='Transformed')

    plt.axhline(0, color='black', lw=1)
    plt.axvline(0, color='black', lw=1)
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()

# Define a simple Scaling Matrix
A_scale = np.array([[2, 0],
                    [0, 0.5]])

plot_transformation(A_scale, "Scaling: Stretch X by 2, Compress Y by 0.5")
```

## The Rotation Matrix

```python
theta = np.radians(45) # 45 degrees
c, s = np.cos(theta), np.sin(theta)

# Standard Rotation Matrix formula
R = np.array([[c, -s],
              [s,  c]])

plot_transformation(R, "Rotation: Rotating Space by 45 Degrees")
```

## The Determinant & Singularity

```python
# A matrix that squashes 2D space into a 1D line
A_singular = np.array([[1, 2],
                       [2, 4]])

det = np.linalg.det(A_singular)
print(f"Determinant: {det}")

if np.isclose(det, 0):
    print("This matrix is 'Singular' - it loses information!")

plot_transformation(A_singular, "Singularity: Squashing 2D into 1D")
```
